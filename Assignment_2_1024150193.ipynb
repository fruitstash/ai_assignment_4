```python
import numpy as np

# 1
array_1d = np.arange(10, 51, 5)

print("Array:", array_1d)
print("Shape:", array_1d.shape)
print("Size:", array_1d.size)
print("Data type:", array_1d.dtype)
```

```python
# 2. Generate a Matrix with Random Values
matrix_4x5 = np.random.randint(1, 101, size=(4, 5))
print("\n4x5 Matrix with random values:\n", matrix_4x5)

max_value = np.max(matrix_4x5)
min_value = np.min(matrix_4x5)
max_index = np.argmax(matrix_4x5)
min_index = np.argmin(matrix_4x5)

print("Maximum value:", max_value)
print("Minimum value:", min_value)
print("Index of maximum value (flattened):", max_index)
print("Index of minimum value (flattened):", min_index)

# using unravel_index here...
max_2d_index = np.unravel_index(max_index, matrix_4x5.shape)
min_2d_index = np.unravel_index(min_index, matrix_4x5.shape)

print("Index of maximum value 2d", max_2d_index)
print("Index of minimum value", min_2d_index)
```

```python
# Given a NumPy array of 15 random integers between 0 and 50, find all elements greater than 25 and replace them with -1.
random_array = np.random.randint(0, 51, 15)
print("Original array:", random_array)

random_array[random_array > 25] = -1
print("Array after boolean indexing:", random_array)
```

```python
# Create two 1D NumPy arrays:
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("Array a:", a)
print("Array b:", b)

#element-wise addition, subtraction, multiplication, and division.
addition = a + b
subtraction = a - b


multiplication = a * b
division = a / b

print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

```python
# 6
# a 5x5 matrix with random floats between 0 and 1.
matrix = np.random.rand(5, 5)
print(matrix)

# Compute the mean, median, variance, and standard deviation.
mean_value = np.mean(matrix)
median_value = np.median(matrix)
variance_value = np.var(matrix)
std_dev_value = np.std(matrix)

print("Mean:", mean_value)
print("Median:", median_value)
print("Variance:", variance_value)
print("Standard deviation:", std_dev_value)
```

```python
# 7
# A is 3x2 with values from 1 to 6.
A = np.arange(1, 7).reshape(3, 2)
print("\nMatrix A (3x2):", A)

# B is 2x3 with values from 7 to 12.
B = np.arange(7, 13).reshape(2, 3)
print("\nMatrix B (2x3):", B)

#matrix multiplication (A @ B)
matrix_product = A @ B
print("\nMatrix product (A @ B):", matrix_product)
```

```python
# 8
array_6x6 = np.arange(1, 37).reshape(6, 6)
print("Original 6x6 array:", array_6x6)

array_6x6[1:4, 1:4] = 0
print("after modification:", array_6x6)
```

```python
# 9 np.where
array_where = np.array([5, 10, 15, 20, 25])
print("Original array for np.where:", array_where)

modified_array_where = np.where(array_where > 15, 99, 0)
print("Array after using np.where:", modified_array_where)
```

```python
# 10. Broadcasting
array_4x4 = np.full((4, 4), 10)
print("original 4x4 array:", array_4x4)

subtraction_array = np.array([0, 1, 2, 3])
broadcasted_result = array_4x4 - subtraction_array
print("after broadcasting", broadcasted_result)
```
